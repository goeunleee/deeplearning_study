```
def And(x, y):
  if x == 1 and y ==1:
    return 1
  if x==1 and y == 0:
    return 0
  if x ==0 and y==1:
    return 0
  if x ==0 and y==0 :
    return 0
```

```
x = [(0,0),(0,1),(1,0),(1,1)]

for i in x :
  print("{}=>{}".format(i,And(i[0],i[1])))
```

```
def And1(x1,x2):
  w1,w2,th= 0.2, 0.4, 0.6
  tmp = x1 * w1 + x2 * w2
  # th : 경계값  
  if(tmp <=th):
    return 0
  else :
    return 1

x = [(0,0),(0,1),(1,0),(1,1)]

for i in x :
  print("{}=>{}".format(i,And1(i[0],i[1])))    
```

```
def And2(x1,x2):
  x = np.array([x1,x2])
  w = np.array([0.5,0.5])
  b = -0.7

  tmp = np.sum(w*x)+b
  if tmp<=0:
    return 0
  else :
    return 1

x = [(0,0),(0,1),(1,0),(1,1)]

for i in x :
  print("{}=>{}".format(i,And2(i[0],i[1])))       
```

```
def NAND(x1,x2):
  x = np.array([x1,x2])
  w = np.array([-0.5,-0.5])
  b = 0.7

  tmp = np.sum(w*x)+b
  if tmp<=0:
    return 0
  else :
    return 1


x = [(0,0),(0,1),(1,0),(1,1)]

for i in x :
  print("{}=>{}".format(i,NAND(i[0],i[1])))
```

```
def OR(x1,x2):
  x = np.array([x1,x2])
  w = np.array([0.5,0.5])

  b = -0.4
  
  tmp = np.sum(w*x)+b
  if tmp<=0:
    return 0
  else :
    return 1

x = [(0,0),(0,1),(1,0),(1,1)]

for i in x :
  print("{}=>{}".format(i,OR(i[0],i[1])))    
```

```
def XOR(x1,x2):
  s1 = NAND(x1,x2)
  s2 = OR(x1,x2)
  y = And1(s1,s2)
  return y

x = [(0,0),(0,1),(1,0),(1,1)]

for i in x :
  print("{}=>{}".format(i,XOR(i[0],i[1])))    
```

```
# 활성화 함수 sigmoid 

import numpy as np
import matplotlib.pyplot as plt
import math
def sigmoid(x):
  return 1/(1+np.exp(-x))

x = np.arange(-0.5, 5.0, 0.1)
#y = sigmoid(x)
y = [math.tanh(z) for z in x]
plt.plot(x,y)
#plt.ylim(-0.1, 1.1)
plt.title('tanh')
plt.show() 
```

```
# 경사하강법

x=1 #입력값
y=0 #출력값

# 랜덤한 초기값
w = tf.random.normal([1], mean=0, stddev=1) #w, b 는 최적의 값 .
b = tf.random.normal([1],mean=0, stddev=1)
for i in range(10000):
  output = sigmoid(x * w + 1 * b)
  error = y - output
  w = w+ x *0.1 * error
  b = b + 1 * 0.1 * error 
  if i % 1000 == 999 :
    print(f"{i}, {error}, {output}") 
```

```
import numpy as np
x = np.array([[1,1],[1,0],[0,1],[0,0]]) # 입력값
y = np.array([[1],[1],[1],[0]])

w = tf.random.normal([2],mean=0, stddev=1) #가중치 : 랜덤한 시작값
b = tf.random.normal([1],mean=0, stddev=1) #편향 : 랜덤한 시작값

for i in range(2000):
  error_sum = 0

  for j in range(4): #입력항목이 4개
     sum = np.sum(x[j]*w) # [1,1] * [0.4 ,0.5]
     output = sigmoid(sum+1*b)
     error = y[j][0] - output
     w = w + x[j] * 0.1 * error
     b = b + 1 * 0.1 * error
  
  if i % 200 == 199 :
    print(f"{i}, {error}, {output}")    
```

```
x = np.array([[1,1],[1,0],[0,1],[0,0]]) # 입력값
y = np.array([[0],[1],[1],[0]])

model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=2, activation='sigmoid', input_shape=(2,)),
    tf.keras.layers.Dense(units=1, activation='sigmoid')
])


model.summary()
```

```
#최적화 확룰적 경사하강법(SGD), 에러처리방법: mse(평균제곱오차)
model.compile(optimizer=tf.keras.optimizers.SGD(lr=0.3), loss="mse")
```

```
history = model.fit(x,y,epochs=2000, batch_size=1, verbose=2)
```

```
plt.plot(history.history['loss'])
```
